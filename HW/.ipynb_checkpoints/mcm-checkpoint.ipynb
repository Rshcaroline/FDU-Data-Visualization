```python
import pandas as pd
import numpy as np
import os

cty = pd.read_csv("location.csv")
b = pd.DataFrame(cty["cty"])
cty = list(cty["cty"])

path = "./1"
files = os.listdir(path=path)
files.remove('.DS_Store')

for yr in range(2008, 2016, 2):
    sum = [0]*50
    for file in files:
        lt = []
        for i in cty:
            a = pd.read_csv(os.path.join(path, file), index_col=0, skiprows=4)
            lt.append(a[str(yr)][i])
        lt = lt/np.linalg.norm(lt)
        sum += lt
        # b[" ".join(file.split(".")[1:3])] = np.array(lt)
        
    sum /= max(sum)
    b["SUM_" + str(yr)] = np.array(sum)
b.to_csv("non-geo-08to14.csv")
```

```python
import pandas as pd
import numpy as np
import csv

a = pd.read_csv('language.csv', header=None, names=["lan", "1st", "2nd", "1st_c", "2nd_c", "tot_1", "tot_2", "yr"])
c = pd.read_csv('pop.csv', index_col=0, skiprows=4)

b = []
for i in range(0, len(a)):
    yr = a["yr"][i]   # use language year's population
    for j in a["1st"][i].split("; "):
        a["tot_1"][i] += c.loc[j, str(yr)]
    for j in a["2nd"][i].split("; "):
        a["tot_2"][i] += c.loc[j, str(yr)]

    for j in a["1st"][i].split("; "):
        p = c.loc[j, str(yr)]/a["tot_1"][i]
        if a["1st_c"][i]:
            b.append([a["lan"][i], j, 1, p*a["1st_c"][i], 0, 0, a["yr"][i]])
    for j in a["2nd"][i].split("; "):
        p = c.loc[j, str(yr)]/a["tot_2"][i]
        if a["2nd_c"][i]:
            b.append([a["lan"][i], j, 2, 0, p*a["2nd_c"][i], 0, a["yr"][i]])

b = pd.DataFrame(b, columns=["lge", "cty", "order", "1st", "2nd", "3rd", "yr"])

b.to_csv("ini_data_new.csv")
```

```python
#  ============================== geo

a = pd.read_csv('language.csv', header=None, names=["lan", "1st", "2nd", "1st_c", "2nd_c", "tot_1", "tot_2", "yr"])

year = 2010

lan_to_cty = {}    # {'Mandarin Chinese': ['China', 'Singapore', 'Vietnam', 'Malaysia', 'Thailand']}
for i in range(0, len(a)):
    if a["yr"][i] == year:
        lan = a["1st"][i].split("; ")
        # lan.extend(a["2nd"][i].split("; "))  # only consider first language
        lan_to_cty[a["lan"][i]] = lan
    
a = pd.read_csv("geo-non-pop-10to14.csv", index_col=0)
cty_to_idx = {}

for i in range(0, len(a)):
    col = a.columns.tolist()
    cty = a.index.tolist()
    val = [a[c][i] for c in col]
    cty_to_idx[cty[i]] = val
    
a = pd.read_csv("ini_data_new.csv", index_col=1)
a.drop(a.columns[[0]], axis=1, inplace=True)    # remove number column

lan = a.index.tolist()
lan_cty_pop = {}

for i in lan:
    lan_cty_pop[i] = {}
for i in range(0, len(a)):
    if a["order"][i] == 1 and a["yr"][i] == year:
            lan_cty_pop[lan[i]][a["cty"][i]] = a["1st"][i]

# for j in range(0, 5):
j = 0
lan_s = {}
for i in lan_to_cty.keys():   # i = Mandarin Chinese
    cty = lan_to_cty[i]       # ['China', 'Singapore']
    GI = [cty_to_idx[i][j] for i in cty]    # 2010 data [0.61961279899999999, 0.55070235600000006]
    s = [GI[n] for n in range(0, len(cty))]   # *lan_cty_pop[i][cty[n]]
    lan_s[i] = np.sum(s)
 
with open('s_' + str(j + 10) + '.csv', 'w') as csv_file:   # 's_' + str(j + 10) + '.csv'
    writer = csv.writer(csv_file)
    for key, value in lan_s.items():
       writer.writerow([key, value])
```

```python
import pandas as pd
from pyecharts import Map
from pyecharts import Style

data = pd.read_csv('ini_data_new.csv')

style = Style(
    title_top="#fff",
    title_pos = "center",
    width=1000,
    height=500,
    background_color="#ffffff",
)

style_geo = style.add(
    is_label_show=False,
    is_roam=False,
    is_legend_show=True,
    maptype="world",
    visual_range=[0, 800],
    visual_range_color=['#6495ED','#4169E1','#0000FF', '#000080']
)

l1 = data[data.order==1]
l2 = data[data.order==2]

attr1=list(l1.cty)
value1=list(l1["1st"])
attr2=list(l2.cty)
value2=list(l2["2nd"])
worldmap = Map("", **style.init_style)
worldmap.add("native language", attr1, value1,  is_visualmap=True, visual_text_color='#000',  is_map_symbol_show=False, **style_geo)
worldmap.add("second language", attr2, value2,  is_visualmap=True, visual_text_color='#000',  is_map_symbol_show=False, **style_geo)
worldmap
```
