```python
# export to python (worldCupAnim_soccerPowerIndex.py) and run using vscode
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

#import matplotlib.animation as animation   # can not use animation in notebook

data_prefix = "../datasets/data/"

data_src = "world-cup-predictions/wc-20140609-140000.csv"

dataReal = data_prefix+data_src

pdata = pd.read_csv(dataReal,sep=",")

minLimit = 0
maxLimit = 31

countryList = pdata["country"].values[minLimit:maxLimit]
spiScore = pdata["spi"].values[minLimit:maxLimit]

x_pos = list(range(len(countryList)))
fig, ax = plt.subplots(figsize=(9,6))
ax.grid()

ax.set_ylim(min(spiScore), max(spiScore))
ax.set_xlim(0, len(x_pos))

line, = ax.plot([], [], lw=1)
xdata, ydata = [],[]

# def animate(i):
# 	print( "append_line ",i )
# 	ax.annotate(countryList[i] + ":" + str(spiScore[i]), (i, spiScore[i] ))
# 	line.set_data(x_pos[0:i+1], spiScore[0:i+1])
# 	return line,
#anim = animation.FuncAnimation(fig,animate,len(x_pos),repeat=False,interval=200,blit=False)
for i in range(len(x_pos)):
    ax.annotate(countryList[i] + ":" + str(spiScore[i]), (i, spiScore[i] ))

ax.set_ylim(60,100)
ax.set_xlim(-2,35)
line.set_data(x_pos[:], spiScore[:])

plt.show()
```
