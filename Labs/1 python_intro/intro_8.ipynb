## Classes and Objects

  **python中的类遵循面向对象的理念来设计，这种理念通常有下面一些特点**
  
* Class
    一种用户定义的包含属性，方法，成员变量的一个数据结构的集合，通常面向某种抽象的概念。
* Class Vairables
    为类的所有实例共享的变量。
* Data member
    在类和类的实例中持有数据的变量。
* Function Overriding
    对一个函数的不同实现方法，在使用的时候可以通过调用着的对象类型和所附加的参数类型来决定调用哪种实现。
* Instance variables
    类中的一个变量，但在实例化后，完全只从属于该实例。
* Inheritance
    类的继承的特性
* Object
    根据类定义的实例化，就是一个对象
* Method
    类中定义的函数方法。
* Operator overrding
    类似于函数重载，一种对于操作符含义的再定义.
    
    
    
    
**下面通过具体的代码介绍类和对象的知识**

```python
#------类的定义------
class Employee:#一个简单的类
   'Common base class for all employees'#通常在这里进行类的简单说明
   empCount = 0#类变量，为所有实例化后的对象所共有

   def __init__(self, name, salary):#特殊函数，用于实例化一个对象
      self.name = name
      self.salary = salary
      Employee.empCount += 1
   
   def displayCount(self):
     print ("Total Employee %d" % Employee.empCount)

   def displayEmployee(self):
      print ("Name : ", self.name,  ", Salary: ", self.salary)
#------函数的定义------


# "This would create first object of Employee class"
emp1 = Employee("Zara", 2000)
# "This would create second object of Employee class"
emp2 = Employee("Manni", 5000)


emp1.displayEmployee()#调用类的成员函数
emp2.displayEmployee()#调用类的成员函数
print ("Total Employee %d" % Employee.empCount)

print("\n\n\n\n\n")

#-------类的属性-------
emp1.age = 7  # Add an 'age' attribute.
# emp1.age = 8  # Modify 'age' attribute.
# del emp1.age  # Delete 'age' attribute.

print(hasattr(emp1, 'age'))    # Returns true if 'age' attribute exists
print(getattr(emp1, 'age'))   # Returns value of 'age' attribute
setattr(emp1, 'age', 8) # Set attribute 'age' at 8
delattr(emp1, 'age')    # Delete attribute 'age'
#-------类的属性-------

print("\n\n\n\n\n")

#-------类的基本属性-------
print ("Employee.__doc__:", Employee.__doc__)#类的说明
print ("Employee.__name__:", Employee.__name__)#类的名称
print ("Employee.__module__:", Employee.__module__)#类的模块名
print ("Employee.__bases__:", Employee.__bases__)#类的所有几基类，会返回一个元组
print ("Employee.__dict__:", Employee.__dict__)#返回一个字典，其中是类的命名空间
#-------类的基本属性-------

print("\n\n\n\n\n")

#-------对象的销毁-------
#通常基础数据的销毁是python解释器自己内部通过引用计数完成的
a = 40      # Create object <40>
b = a       # Increase ref. count  of <40> 
c = [b]     # Increase ref. count  of <40> 

del a       # Decrease ref. count  of <40>
b = 100     # Decrease ref. count  of <40> 
c[0] = -1   # Decrease ref. count  of <40> 此时a的内存才真正被释放

#类实例的销毁跟上述过程类似
class Point:
   def __init( self, x=0, y=0):
      self.x = x
      self.y = y
   def __del__(self):
      class_name = self.__class__.__name__
      print (class_name, "destroyed")

pt1 = Point()
pt2 = pt1
pt3 = pt1
print (id(pt1), id(pt2), id(pt3)) # prints the ids of the obejcts
del pt1
del pt2
del pt3#到这里类实例才真正被销毁
#-------对象的销毁-------

print("\n\n\n\n\n")


#-------类的继承-------
class Parent:        # define parent class
   parentAttr = 100
   def __init__(self):
      print ("Calling parent constructor")

   def parentMethod(self):
      print ('Calling parent method')

   def setAttr(self, attr):
      Parent.parentAttr = attr

   def getAttr(self):
      print ("Parent attribute :", Parent.parentAttr)

class Child(Parent): # define child class
   def __init__(self):
      print ("Calling child constructor")

   def childMethod(self):
      print ('Calling child method')

c = Child()          # instance of child
c.childMethod()      # child calls its method
c.parentMethod()     # calls parent's method
c.setAttr(200)       # again call parent's method
c.getAttr()          # again call parent's method
#-------类的继承-------

print("\n\n\n\n\n")

#-------类方法重载-------
class Parent:        # define parent class
   def myMethod(self):
      print ('Calling parent method')

class Child(Parent): # define child class
   def myMethod(self):
      print ('Calling child method')

c = Child()          # instance of child
c.myMethod()         # child calls overridden method
#-------类方法重载-------

print("\n\n\n\n\n")

#-------操作符重载-------
class Vector:
   def __init__(self, a, b):
      self.a = a
      self.b = b

   def __str__(self):
      return 'Vector (%d, %d)' % (self.a, self.b)
   
   def __add__(self,other):
      return Vector(self.a + other.a, self.b + other.b)

v1 = Vector(2,10)
v2 = Vector(5,-2)
print (v1 + v2)
#-------操作符重载-------

print("\n\n\n\n\n")


#-------隐藏类的属性-------
#当我们希望某些类的变量不被外界访问到时

class JustCounter:
   __secretCount = 0
  
   def count(self):
      self.__secretCount += 1
      print (self.__secretCount)

counter = JustCounter()
counter.count()
counter.count()
print (counter.__secretCount)
#-------隐藏类的属性-------
```
